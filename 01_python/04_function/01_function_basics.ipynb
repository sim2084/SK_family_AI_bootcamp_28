# 함수 기초 (Function Basics)
- 함수 :  특정 작업을 수행하는 코드의 묶음. 한 번 정의하면 언제든 재사용 가능하다.

```python
# 가장 기본적인 함수와 호출의 예시
def basic_function():          #정의
    print('Hello World!')

basic_function()    #호출
basic_function()
```

## 매개변수(parameter)와 인자(argument)
- 매개변수 : 함수 선언부의 변수 (공간)
- 인자 : 함수 호출부에서 전달하는 값 (실체)

```python
def add(a,b):                       # a, b가 parameter
    print(f'a = {a}, b = {b}')
    return a + b

print(f'결과1 : {add(10,20)}')       # 10, 20이 argument
print(f'결과2 : {add('Hello','World')}')
```

## 위치 인자 vs 키워드 인자
- 위치 인자: 작성된 순서대로 전달
- 키워드 인자: name=value 형태로 이름을 저장하여 전달

```python
def test(a,b):
    print(f'a={a}, b={b}')

test(10,20)         #위치 인자
test(b=200,a=100)   #키워드 인자 (순서 상관없음)
test(10,b=20)       #혼용 시 위치 인자가 항상 먼저 와야함
```

## 매개변수 기본 값
- 인자를 생략했을 때 사용할 기본 값을 지정할 수 있다.
- 주의: 기본 값이 있는 매개변수는 반드시 오른쪽에 위치

```python
def greeting(name, greet='처음 뵙겠습니다.'):
    print(f'{name}: {greet}')

greeting('다람쥐')
greeting('호랑이','안녕?')
```

## 반환 값
- 함수 실행 후 결과 값을 호출 부로 돌려준다.
- 여러 개의 값을 한 번에 반환할 수도 있다. (내부적으로 튜플이 활용된다.)

```python
def calc(a,b):
    return a+b, a-b, a*b, a/b   # 4개의 값을 담은 튜플을 변환

res = calc(10,2)
print(f'전체 결과: {res}')
p,m,mul,div = calc(10,2)    #언패킹하여 각각의 변수 담기
print(f'덧셈: {p}, 나눗셈: {div}')
```
