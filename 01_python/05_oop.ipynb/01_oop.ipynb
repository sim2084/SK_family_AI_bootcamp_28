# OOP (Object Oriented Programming) : 객체 지향 프로그래밍
객체(메모리 조각)를 중심으로 프로그램을 작성하는 방법론
- 클래스는 관련 데이터(속성)와 기능(메소드)을 하나의 묶음으로 만드는 설계도 => 구조화
- 클래스를 작성 후에 이를 바탕으로 n개의 객체(instance)를 생성할 수 있다. => 재사용성
- 각 객체는 서로 다른 메모리 공간에 존재하므로 다른 객체에 영향을 주지 않는다 => 독립성

```python
# 클래스 (객체의 설계도)
class Person:

    def __init__(self,name,age):
        """생성자 함수 (객체의 속성 초기화를 담당)"""
        # self -> 현재 객체 (메모리 주소)
        self.name = name
        self.age = age

    def greeting(self):
        print(f'안녕하세요. 저는 {self.age}살 {self.name}입니다.')

# 클래스와 객체(instance)
hong = Person('홍길동',33)
sinsa = Person('신사임당',40)

print(hong, type(hong))
print(sinsa, type(sinsa))

print(hong.name,hong.age)
hong.greeting()

print(sinsa.name,sinsa.age)
sinsa.greeting()
```

## 속성과 메소드
- 클래스 속성: 인스턴스 간에 공유 될 수 있는 속성
- 클래스 메소드: 인스턴스 간에 공유 될 수 있는 메소드
- 인스턴스 속성: 인스턴스마다 달라지는 속성
- 인스턴스 메소드: 인스턴스 속성에 대해 처리하는 메소드

```python
class SKNetworkStudent:
    # 클래스 속성
    course_name = 'Family Ai Camp'

    # 생성자
    def __init__(self,nth,name):
        # 인스턴스 속성
        self.nth = nth
        self.name = name

    # 인스턴스 메소드
    def greeting(self):
        print(f'안녕하세요. {self.nth}기 {self.name}입니다.')

    #클래스 메소드
    # @ <- 데코레이터 : 기존 함수를 수정하지 않고 새로운 기능이나 특성을 부여하는 문법
    @classmethod
    def good_morning(cls):      # 클래스 객체 사용 가능
        print(f'{cls.course_name} 학생 여러분, 좋은 아침입니다.')

    # 클래스와 논리적으로 연관은 있지만 독립적인 유틸리티 기능을 가진 함수
    @staticmethod
    def good_bye(name):
        print(f'{name}님, 오늘도 수고하셨습니다.')

# 인스턴스 생성
kim = SKNetworkStudent(28, '김철수')
lee = SKNetworkStudent(27, '이민수')

print(kim, type(kim))
print(lee, type(lee))

# 클래스 객체
print(SKNetworkStudent, type(SKNetworkStudent))

# 클래스의 속성과 메소드
print(SKNetworkStudent.course_name)
SKNetworkStudent.good_morning()

# [주의] 인스턴스를 통해 클래스 변수를 수정하려고 한다면?
kim.course_name = '빅데이터 과정'
print(kim.course_name)                  # 인스턴스 변수 course_name이 새로 생김
print(SKNetworkStudent.course_name)     # 클래스 변수 값은 그대로
print(lee.course_name)                  # 클래스 변수 값 참조
```

# 객체 지향의 원칙
1. 캡슐화 : 속성/기능(메소드)을 묶어서 선언하고, 속성 직접 접근은 막고 메소드를 통한 사용만 허용하는 것
2. 상속 : 부모 클래스에 선언 된 속성/기능을 자식 클래스에서 선언 없이 사용하는 것(재사용성)
3. 다형성 : 상속에 기반한 기술로 여러 자식 인스턴스를 부모 타입으로 제어하는 것
4. 추상화 : 공통적, 보편적인 속성/기능을 추출해 내는 것
