# 주석 (Comment)
**주석** : 컴퓨터가 해석하지 않고 무시하는 구문으로 코드 설명 또는 특정 코드의 실행을 방지할 때 사용한다.

````python
# 한 줄 주석 (ctrl + /)
print('이 문장은 실행함')
# print('이 문장은 실행 안함')
````

````python
# 여러 줄 주석 : 정보나 긴 설명을 작성 시 사용
# 작성자 : 홍길동
# 작성일 : 2026-02-27
# 내용 : 파이썬 변수 기초 실습
print('Hello Python!')
````

# 변수 (Variable)
데이터를 저장하는 **메모리 공간**의 이름

## 변수 사용 이유
- 유지보수
- 재사용
- 가독성

````python
# '=' 연산자 : 오른쪽 값을 왼쪽 변수에 대입(할당)
team_name = "ohgiraffers"
print(team_name)

# 값 변경 : 기존 데이터 위에 새로운 값이 덮어 씌워짐
team_name = "오지라퍼스"
print(team_name)

# 값 복사 : 변수에 담긴 값을 다른 변수에 할당 가능
my_team = team_name
print(my_team)
````

# 변수 이름 짓기 (Naming Rule)
1) 필수 규칙 (위반하면 에러)
- 사용 가능 문자: 영문 대소문자, 숫자, 언더바
- 숫자로 시작 불가
- 특수문자/공백 불가
- 예약어 사용 불가
2) 권장 관례 (가독성을 위한 약속)
- snake_case : 단어 사이를 언더바로 연결
- 직관적 명명
- 한글 사용 가능하지만 인코딩/호환성 이슈 있을 수 있으므로 지양

````python
# 대소문자 구문
X = 1
x = 2
print(X, x)

# 숫자 사용
user1 = "홍길동"
user_2 = "유관순"
# 3user = "안됨"
print(user1, user_2)

# 스네이크 케이스
teamname = "오지라퍼스"
team_name = "ohgiraffers"

# 한글 값 사용
팀명 = "오지라퍼스"
print(팀명)

# 에러 케이스
# 1st_user = '숫자 시작 안됨'
# user-name = '하이픈 안됨'
# user email = '공백 안됨'
# class = '예약어 안됨'

# 예약어 목록 확인
import keyword
print("파이썬 예약어 목록")
print(keyword.kwlist)

# 의미 있는 이름 사용
a = True
is_user_logged_in = True
````
